#Splitting Train, Test and Pool data

```python
import numpy as np
from sklearn.model_selection import train_test_split


train_ratio = 0.2
pool_ratio = 0.6
test_ratio = 1 - train_ratio - pool_ratio



labels = data["Is Fraud?"]
data = data.drop("Is Fraud?", axis=1)


# First split: 60% for training, 40% for remaining data
data_train, data_remaining, labels_train, labels_remaining = train_test_split(data, labels, test_size=(1 - train_ratio), random_state=42)

# Second split
data_pool, data_test, labels_pool, labels_test = train_test_split(data_remaining, labels_remaining, test_size=0.375, random_state=42)

# Now, data_train, labels_train contain 20% of the data
# data_pool, labels_pool contain 50% of the data
# data_test, labels_test contain 30% of the data
```

```python
data_train["Is Fraud?"] = labels_train
data_train
```

```python
data_pool["Is Fraud?"] = labels_pool
data_pool
```

```python
data_test["Is Fraud?"] = labels_test
data_test
```

Handling Missing Values

```python
from sklearn.preprocessing import LabelEncoder

class preprocessor(object):
  def __init__(self, df):
    self.df = df
  def handle_missing_values(self):
    nan_cols = ["Merchant State","Zip"]
    self.df = self.df.drop("Errors?", axis = 1)
    self.df[nan_cols] = self.df[nan_cols].fillna(self.df[nan_cols].mode().iloc[0])

  def handle_amount(self):
    self.df['Amount'] = pd.to_numeric(self.df['Amount'].str.replace('$', ''), errors='coerce')

  def handle_label(self):
    le = LabelEncoder()
    self.df['Is Fraud?'] = le.fit_transform(self.df['Is Fraud?'])

  def handle_time(self):
    self.df['Hours'] = self.df['Time'].str.split(':').str[0].astype(int)
    self.df['Minutes'] = self.df['Time'].str.split(':').str[1].astype(int)
    self.df['Total_minutes'] = self.df['Hours'] * 60 + self.df['Minutes']
    self.df = self.df.drop("Time", axis = 1)
    self.df = self.df.drop("Hours", axis = 1)
    self.df = self.df.drop("Minutes", axis = 1)

  def handle_categorical(self):
    categorical_columns = self.df.select_dtypes(include=['object']).columns.tolist()
    le = LabelEncoder()

    for col in categorical_columns:
        self.df[col + '_encoded'] = le.fit_transform(self.df[col])

    self.df = self.df.drop(categorical_columns, axis=1)
  def run(self):
    self.handle_missing_values()
    self.handle_amount()
    self.handle_label()
    self.handle_time()
    self.handle_categorical()
    print(self.df.info())
    print("--------------------------")
    return self.df


```

```python
train_preprocessor = preprocessor(data_train)
data_train = train_preprocessor.run()
```

```python
pool_preprocessor = preprocessor(data_pool)
data_pool = pool_preprocessor.run()
```

```python
test_preprocessor = preprocessor(data_test)
data_test = test_preprocessor.run()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
y_train = data_train["Is Fraud?"]
X_train = data_train.drop("Is Fraud?", axis=1)

y_test = data_test["Is Fraud?"]
X_test = data_test.drop("Is Fraud?", axis=1)


rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)


rf_classifier.fit(X_train, y_train)

y_pred = rf_classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')
from sklearn.metrics import f1_score

f1 = f1_score(y_test, y_pred)
print(f'f1_score: {f1* 100}%')
```
